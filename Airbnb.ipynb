```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Reading

```python
df = pd.read_csv("compressed_data.csv")
df.head()
```

# Data Cleaning 

```python
# checking missing value 
df.isnull().sum()
```

```python
df.info()
```

Handling Missing Value

```python
df['last review'] = pd.to_datetime(df['last review'],errors='coerce')
df.info()
```

```python
df.fillna({'review per month':0,'last review':df['last review'].min()} , inplace=True)
df.dropna(subset = ['NAME',"host name"],inplace = True)
df.isnull().sum()
```

```python
df = df.drop(columns=['license','house_rules'],errors='ignore')
df.head()
```

```python
#Remove Dollar sign and convert to float 
df.price = df.price.replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
df.head()
```

Remove Duplicates 

```python
df.drop_duplicates(inplace=True)
df.info()
```

# Display Statistics

```python
df.describe()
```

# Visualization 

What is the distribution of listing Price

```python
plt.figure(figsize = (10,6))
sns.histplot(df.price,bins =50,kde = True, color = 'red')
plt.title("Distribution of Listing Price")
plt.xlabel("Price($)")
plt.ylabel("Frequency")
plt.show()
```

The Histogram shows a fairly even distribution of listing price across different price ranges ,
indicating no particular concentration of listing in specific price range .
The KDE line help visualize this more clearly , confirming that
the data set consist listing with wide variety of prices

How are different room type distributed ?

```python
df['room type'].head()
```

```python
plt.figure(figsize =(8,5))
sns.countplot(x='room type',data=df,color='hotpink')
plt.title("Room Distribution")
plt.xlabel("Room")
plt.ylabel("Count")
plt.show()
```

How are the listing distributed across different neighbourhood 

```python
plt.figure(figsize=(8,5))
sns.countplot(y='neighbourhood group',data = df , color = 'lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title("Number of Listing by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Group")
plt.show()
```

What is the relationship between price and room type 

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type",y="price",hue="room type",data = df , palette='Set1')
plt.title("Price v/s Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price($)")
plt.show()
```

How No. of review changed over time 

```python
df['last review'] = pd.to_datetime(df['last review'])
review_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,6))
review_over_time.plot(kind='line',color='red')
plt.title("No. of review over time")
plt.xlabel('Date')
plt.ylabel('No. of Review')
plt.show()
```
